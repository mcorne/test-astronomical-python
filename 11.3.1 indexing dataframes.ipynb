```python
from pathlib import Path

import pandas as pd
from astropy.table import Table

sfr_file = Path(Path.cwd()) / f"data/catalogs/cosmos_3dhst.v4.1.5.zbest.sfr"
fast_file = Path(Path.cwd()) / f"data/catalogs/cosmos_3dhst.v4.1.5.zbest.fout"
rf_colors_file = Path(Path.cwd()) / f"data/catalogs/cosmos_3dhst.v4.1.master.RF.FITS"
photometry_file = Path(Path.cwd()) / f"data/catalogs/cosmos_3dhst.v4.1.cat.FITS"

sfr_df = pd.read_csv(sfr_file, sep="\s+")  # delim_whitespace=True deprecated
fast_df = pd.read_csv(fast_file, sep="\s+")
rf_colors_df = Table.read(rf_colors_file)
photometry_df = Table.read(photometry_file).to_pandas()
```

```python
df_high_sf = sfr_df.loc[
    (sfr_df.sfr > 100) & (sfr_df.flag == 0),
    ["id", "z_best", "z_type", "sfr", "flag"],
]
print(df_high_sf.head())
print(f"{len(df_high_sf)} galaxies meet criteria.")
```

```python
# merged_df = sfr_df.merge(fast_df, on="id") # not necessary since "merge" can be done on indexes
df_m_sf = fast_df.loc[
    ((sfr_df.sfr > 100) | (fast_df.lmass > 12)) & (sfr_df.flag == 0),
    ["id", "lmass", "metal", "lage"],
]
print(df_m_sf)
```
