```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

with fits.open(Path(Path.cwd()) / f"data/HST_JWST/jwst_f277W_crop.fits") as hdu:
    image = hdu[0].data
    header = hdu[0].header

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image, origin="lower")
```

```python
for key, value in list(header.items())[:5]:
    print(f"{key}: {value}")
```

```python
s = 5

vmin = np.mean(image) - s * np.std(image)
vmax = np.mean(image) + s * np.std(image)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image, vmin=vmin, vmax=vmax, origin="lower")
```

```python
s = 1

vmin = np.mean(image) - s * np.std(image)
vmax = np.mean(image) + s * np.std(image)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image, vmin=vmin, vmax=vmax, origin="lower")
```

```python
s = 1.5
cropped = image[870:1080, 1020:1230]
vmin = np.mean(cropped) - s * np.std(cropped)
vmax = np.mean(cropped) + s * np.std(cropped)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(cropped, vmin=vmin, vmax=vmax, origin="lower", cmap="gray_r")
```

```python
s = 1.5
cropped = image[870:1080, 1020:1230]

vmin = np.mean(cropped) - s * np.std(cropped)
vmax = np.mean(cropped) + s * np.std(cropped)

log_cropped = np.log1p(cropped)

log_vmin = np.log1p(vmin)
log_vmax = np.log1p(vmax)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(log_cropped, vmin=log_vmin, vmax=log_vmax, origin="lower", cmap="gray_r")
```
